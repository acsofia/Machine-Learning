```python
#import library to deal with the dataset
import pandas as pd 
```

```python
#load the dataset
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
#after splitting the input and output, as the input value has categorical data in it, convert in to numeric with get dummies
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
#Split input and out from the data set
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit',
       'State_Florida', 'State_New York']]
independent
```

```python
dependent=dataset[['Profit']]
dependent
```

```python
#now split the Train set and Test set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#now create the model
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='absolute_error',splitter='random')
regressor=regressor.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
#Evaluate the model
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
